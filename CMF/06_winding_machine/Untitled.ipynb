```python
sr = 100
length = 2
t = np.linspace(0, length, sr*length)
x = np.concatenate((np.sin(t[:sr * length //2 ] * 2 * np.pi * 2), 
                    np.sin(t[:sr * length//2] * 2 * np.pi *5)*0.5))

dft = np.fft.fft(x)

N = x.size
k = np.arange(N)
f = sr * k / N


plt.figure(figsize = (18, 6))
plt.subplot(2,1,1)
plt.plot(t, x)
plt.plot(np.abs(dft))
plt.xlim(0,7)
plt.show()


```
